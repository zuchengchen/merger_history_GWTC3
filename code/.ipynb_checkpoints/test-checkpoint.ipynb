```julia
function rungekutta4(f, y0, Î· ) # 4th-order RG method to solve the Eqa for each mode k
    n = length(Î·)
    y = zeros((n, length(y0)))
    y[1,:] = y0
    for i in 1:n-1
        h = Î·[i+1] - Î·[i]
        k1 = f(y[i,:], Î·[i] )
        k2 = f(y[i,:] + k1 * h/2, Î·[i] + h/2 )
        k3 = f(y[i,:] + k2 * h/2, Î·[i] + h/2 )
        k4 = f(y[i,:] + k3 * h, Î·[i] + h )
        y[i+1,:] = y[i,:] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    end
    return y
end
```
