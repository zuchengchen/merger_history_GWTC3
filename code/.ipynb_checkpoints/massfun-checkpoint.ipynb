## test Riotto's mass function

```julia
logks = range(log10(4.5e6*k0/1.0e9),log10(4.5e9*k0),length=300);
ks = 10 .^logks;
```

```julia
function Power(A,α,k)
    kshort = 3.0e8*k0
    klong = kshort/1.0e9
    A*heaviside(kshort - k)*heaviside(k - klong)
end
```

```julia
function σk2_int0(k,A,q,α)
    if k==0
        return 0
    else
        return (4/9)^2.0 /q * (q/k)^4.0 * Window(q,1.0/k)^2.0 * TRD(q,1.0/k)^2.0 * Power(A,α,q)
    end
end


function σk2(k,A,α)
    qmax = log10(3.0e8*k0)
    qmin = log10(qmax/1.0e9)
    σk2_int_log(k,A,q,α) = 10.0^q*σk2_int0(k,A,10.0^q,α) * log(10.0)
    σk2_int1(k,A,q,α) = (qmax-qmin)* σk2_int_log(k,A,qmin+(qmax-qmin)*q,α)
    function integrand(x, f)
        f[1] = σk2_int1(k,A,x[1],α)
    end
    result, err = cuhre(integrand , rtol=1e-4, maxevals=Int(1e9))
    result[1]
end
```

```julia
σs = σk2.(ks,0.001,0)
σtest(k)=extrapolate(interpolate((ks,), σs, Gridded(Linear())), 0)(k)
```

```julia
@time plot(ks/k0,σtest.(ks))
semilogx()
semilogy()
```

```julia
function βmQCD_int0(k,M,A,α,fpbh)
    Mk = 1.4e13/k^2.0
    κ  = 3.3
    γ  = 0.36
#     δc = 0.4531846468157414
    σ  = σtest(k)
#     if σ == 0
#         return 0
#     else
#         term1 = M/Mk
#         term2 = 1/sqrt(2.0*π*σ) * exp(-(δc+(M/Mk/κ)^(1.0/γ))^2.0/(2.0*σ))
#         term3 = 2.8e13*(M/Mk/κ)^(1.0/γ)
#         term4 = k^3.0 *γ *Mk
#         return term1*term2*term3/term4/fpbh/0.84*(2.8e17/M)^0.5/M
#     end
    if σ == 0
        return 0
    else
        term1 = M/Mk
        term2 = 1/sqrt(2.0*π*σ) * exp(-(δc(Mk)+(M/Mk/κ)^(1.0/γ))^2.0/(2.0*σ))
        term3 = 2.8e13*((M/Mk/κ)^(1.0/γ))#-1.0*Mk*γ*Dδc(δc)(Mk))
#         term3 = 2.8e13*((M/Mk/κ)^(1.0/γ)-Dδc(δc)(Mk))
        term4 = k^3.0 *γ *Mk
        result = term1*term2*term3/term4/fpbh/0.84*(2.8e17/M)^0.5/M
        if result>0
            return result
        else
            return 0
        end
    end
end

function βm_int0(k,M,A,α,fpbh)
    Mk = 1.4e13/k^2.0
    κ  = 3.3
    γ  = 0.36
    δc = 0.4531846468157414
    σ  = σtest(k)
    if σ == 0
        return 0
    else
        term1 = M/Mk
        term2 = 1/sqrt(2.0*π*σ) * exp(-(δc+(M/Mk/κ)^(1.0/γ))^2.0/(2.0*σ))
        term3 = 2.8e13*(M/Mk/κ)^(1.0/γ)
        term4 = k^3.0 *γ *Mk
        return term1*term2*term3/term4/fpbh/0.84*(2.8e17/M)^0.5/M
    end
end
```

```julia
function Pm(M,A,α,fpbh)
    kmax = log10(kMass(1e-15)*5)
    kmin = log10(1.0e5)
    βm_int1_log(k,M,A,α,fpbh) = 10.0^k*βm_int0(10.0^k,M,A,α,fpbh) * log(10.0)
    βm_int1(k,M,A,α,fpbh) = (kmax-kmin)* βm_int1_log(kmin+(kmax-kmin)*k,M,A,α,fpbh)
    
#     kmax = 4.5e7*k0
#     kmin = kmax/1.0e9
#     βm_int1(k,M,A,α,fpbh) = (kmax-kmin)* βm_int0(kmin+(kmax-kmin)*k,M,A,α,fpbh)
    function integrand(x, f)
        f[1] = βm_int1(x[1],M,A,α,fpbh)
    end
    result, err = cuhre(integrand,rtol=1e-3,minevals=1e3,maxevals=Int(1e9))
    result[1]
end

function PmQCD(M,A,α,fpbh)
    kmax = log10(kMass(1e-15)*5)
    kmin = log10(1.0e5)
    βm_int1_log(k,M,A,α,fpbh) = 10.0^k*βmQCD_int0(10.0^k,M,A,α,fpbh) * log(10.0)
    βm_int1(k,M,A,α,fpbh) = (kmax-kmin)* βm_int1_log(kmin+(kmax-kmin)*k,M,A,α,fpbh)
    
#     kmax = 4.5e7*k0
#     kmin = kmax/1.0e9
#     βm_int1(k,M,A,α,fpbh) = (kmax-kmin)* βm_int0(kmin+(kmax-kmin)*k,M,A,α,fpbh)
    function integrand(x, f)
        f[1] = βm_int1(x[1],M,A,α,fpbh)
    end
    result, err = cuhre(integrand,rtol=1e-3,minevals=1e3,maxevals=Int(1e9))
    result[1]
end
```
